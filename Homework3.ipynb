\section*{Problem 1}

\subsection*{Part i}

Determine whether one of the alternatives probabilistically dominates the the other alternative

\subsection*{Part i - Solution}

The following code plots the probability functions to be compared.  In this case, neither function dominates the other since the functions cross each other around $x_4$

```python
# Import numpy for array functions
import numpy as np

# Define probablility of each outcome in a'
P_ap     = np.array([.1,.25,.2,.2,.15,.1])
P_ap_cdf = np.array([1.,.9,.65,.45,.25,.1])

# Define probablility of each outcome in a"
P_app = np.array([.2,.2,.2,.1,.1,.2])
P_app_cdf = np.array([1.,.8,.6,.4,.3,.2])

# Define utility of each outcome
u = np.array([.1,.2,.5,.7,.85,1.])

# Import matplotlib for plotting
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6])

plt.plot(x,P_ap_cdf,label="a'")
plt.plot(x,P_app_cdf,label='a"')
plt.legend()
plt.xlabel('x')
plt.ylabel('1 - CDF')
plt.title('1 - cumulative distribution function for alternatives')
plt.show()
```

\subsection*{Part ii}

Calculate the expected utility for the two alternatives. Which alternative should the decision maker choose based on the available information?

\subsection*{Part ii - Solution}

The expected utilities of the two alternatives are the same.  This information, combined with part 1, lead the decision maker to think that both alternative are equivalent.

```python
u_ap  = np.dot(P_ap,u)
u_app = np.dot(P_app,u)

print("Expected utility of a':\t%.4f"%u_ap)
print('Expected utility of a":\t%.4f'%u_app)
```

Calculate the expected utility for the two alternatives. Based on the available information, which alternative should the decision maker choose?

\subsection*{Part i - Solution}

Based on the code below, a" expresses a higher certainty equivalent and should be selected.

```python
from scipy import integrate

def p_p(x):
    mu = 1.
    sigma = .5
    return (np.exp(-.5*((x-mu)/sigma)**2))/(sigma*np.sqrt(2*np.pi))

def p_pp(x):
    return (1./3.)

def u(x):
    return 1. - np.exp(-.2*x)

# Alternative a'
E_ap = integrate.quad(lambda x: p_p(x)*u(x),0,10)

# Alternative a''
E_app = integrate.quad(lambda x: p_pp(x)*u(x),0,3)

print("The certainty equivalent of a':\t%.4f"%E_ap[0])
print('The certainty equivalent of a":\t%.4f'%E_app[0])
```

\section*{Poblem 4}

\begin{equation}\notag
\ r(x) = -\frac{u''(x)}{u'(x)}
\end{equation}

\subsection*{Part i - Solution}

\begin{equation}\notag
\ u_1(x) = -e^{-2x}-2e^{-3x}
\end{equation}
\begin{equation}\notag
\ u_1'(x) = 2e^{-2x}+6e^{-3x}
\end{equation}
\begin{equation}\notag
\ u_1''(x) = -4e^{-2x}-18e^{-3x}
\end{equation}
\begin{equation}\notag
\ r_1(x) = -\frac{-4e^{-2x}-18e^{-3x}}{2e^{-2x}+6e^{-3x}}
\end{equation}

Since we are writing code to solve this function, there is no need to simplify the terms further.

The code in this next section shows that the user is risk averse as the function $r_1(x)$ is always above the threshold of indifference at $r(x) = 0$. As the value of x increases, the decision becomes decreasingly risk averse and reaches a constant asymptote around a value of $r=2$.

```python
def r(x):
    return -1*(-4*np.exp(-2*x)-18*np.exp(-3*x))/(2*np.exp(-2*x)+6*np.exp(-3*x))

x = np.linspace(-10,10,101)
y = r(x)

plt.plot(x,y)
plt.title('Risk of utility function u_1')
plt.xlabel('x')
plt.ylabel('Risk')
plt.show()
```

\subsection*{Part ii - Solution}

\begin{equation}\notag
\ u_2(x) = x + 2 \, ln(x+1)
\end{equation}
\begin{equation}\notag
\ u_2'(x) = 1 + \frac{2}{x+1}
\end{equation}
\begin{equation}\notag
\ u_2''(x) = -\frac{2}{(x+1)^2}
\end{equation}
\begin{equation}\notag
\ r_2(x) = -\frac{-\frac{2}{(x+1)^2}}{1 + \frac{2}{x+1}}
\end{equation}
\begin{equation}\notag
\ r_2(x) = \frac{2}{(x+1)^2 + 2(x+1)}
\end{equation}

One extra simplification step was taken to create an easier function to interpret.

The code in this next section shows that the user is generally risk averse as the function $r_2(x)$ is above zero at all values outside of $-3 \leq x \leq -1$.  This may not prove to matter depending on the possible interval of x (for example, if $x>0$ then the decision maker is always risk averse, approaching risk neutral.

This function proves to asymptote to $r_2(x) = 0$ at $x= \pm \infty$. As the value of x increases, the decision maker becomes decreasingly risk averse and will eventually become risk neutral.

```python
def r(x):
    return 2/((x+1)**2+2*(x+1))

x = np.linspace(-10,10,1001)
y = r(x)

plt.plot(x,y)
plt.title('Risk of utility function u_2')
plt.xlabel('x')
plt.ylabel('Risk')
plt.show()
```
